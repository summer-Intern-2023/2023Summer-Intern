```python
import urllib.parse
import urllib.request
import json

# Base URL for the API endpoint
base_url = "http://py4e-data.dr-chuck.net/json?"

# Parameters for the API request
parameters = {
    "address": "Stonehill College",
    "key": "YOUR_API_KEY"  # Replace with your API key
}

# Encode the parameters
url = base_url + urllib.parse.urlencode(parameters)

try:
    # Make the API request
    response = urllib.request.urlopen(url)
    data = response.read().decode()

    # Parse the JSON response
    json_data = json.loads(data)

    # Extract the place_id
    place_id = json_data["results"][0]["place_id"]

    # Print the place_id
    print("Place ID:", place_id)

except urllib.error.HTTPError as e:
    print("HTTP Error:", e.code, e.reason)

except urllib.error.URLError as e:
    print("URL Error:", e.reason)

except json.JSONDecodeError as e:
    print("JSON Decode Error:", e.msg)

except KeyError:
    print("Error: Unable to retrieve place_id from the response.")
```

```python
import urllib.parse
import urllib.request
import json

# Base URL for the API endpoint
base_url = "http://py4e-data.dr-chuck.net/json?"

# Prompt for location
location = input("Enter location: ")

# Parameters for the API request
parameters = {
    "address": location,
    "key": "YOUR_API_KEY"  # Replace with your actual API key
}

# Encode the parameters
url = base_url + urllib.parse.urlencode(parameters)

try:
    # Make the API request
    print("Retrieving", url)
    response = urllib.request.urlopen(url)
    data = response.read().decode()

    # Print the number of characters retrieved
    print("Retrieved", len(data), "characters")

    # Parse the JSON response
    json_data = json.loads(data)

    # Extract the first place_id
    place_id = json_data["results"][0]["place_id"]

    # Print the place_id
    print("Place ID:", place_id)

except urllib.error.HTTPError as e:
    print("HTTP Error:", e.code, e.reason)

except urllib.error.URLError as e:
    print("URL Error:", e.reason)

except json.JSONDecodeError as e:
    print("JSON Decode Error:", e.msg)

except KeyError:
    print("Error: Unable to retrieve place_id from the response.")
```
